## Types of joining the Dataframes

Concate : It is used to join two dataframe
parameters of the concate    
ignore_index :- It is used to take indexing in contiunous form
axis = 0 == rows
axis =1 == columns
join == inner == axis=0  it is taking the common feature data from the both dataframe according to the rows 
join == inner == axis=1 it is taking the all records according to the Columns 

Join == outer == axis = 0 giving all the data according to the rows
join == outer == axis =1 giving all the data according to the columns

## merge()

```python
display(d1,d2)
```

```python
pd.merge(d1,d2)
```

```python
pd.merge(d1,d2, how='outer')
```

```python
d2 = pd.DataFrame({'ASCII':np.arange(74,79), 'Char1':[chr(i) for i in range(74,79)]})
d2
```

```python
pd.merge(d1,d2, how='inner')
```

```python
d1.rename(columns={'ASCII':'Num'},inplace=True)
```

```python
d1.merge(d2,left_on='Num', right_on='ASCII') # d1, d2
```

```python
d2.merge(d1,left_on='ASCII', right_on='Num') 
```

```python
display(d1,d2)
```

```python
d1.merge(d2,left_on='Num', right_on='ASCII', how='left')
# how =left join it will give u a comman data as well as left data 
```

```python
d1.merge(d2,left_on='Num', right_on='ASCII', how='right')
# how =right join it will give u a comman data as well as right data 
```

```python
d2.merge(d1,left_on='ASCII', right_on='Num', how='right')
```

```python
d2.merge(d1,left_on='ASCII', right_on='Num', how='left')
```
