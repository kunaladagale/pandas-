## loc function

It will access actual index and columns.

```python
df
```

```python
df[2:5] # 2,3,4
```

```python
df[3:7]  # 3,4,5,6   n to n-1
```

```python
df.loc[3:7] # 3,4,5,6,7     n to n records 
```

```python
df[1:10]   # 1 ,2,3,4 .... 9 
```

```python
df.loc[1:10]    #  n to n records
```

```python
df.loc[:]
```

```python
df.loc[:,['PetalLengthCm']]   # df.loc[use for the rows, use for the column]
```

```python
df.loc[:,['PetalLengthCm','PetalWidthCm']]
```

```python
df.loc[2:5, ['PetalLengthCm','PetalWidthCm']]
```
